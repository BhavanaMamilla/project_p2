```python

# Ensure the 'Date_and_Time_When_Order_Was_Placed' column is in datetime format
df['Date_and_Time_When_Order_Was_Placed'] = pd.to_datetime(df['Date_and_Time_When_Order_Was_Placed'], errors='coerce')

# Filter out rows where 'Date_and_Time_When_Order_Was_Placed' is null
df_filtered = df[df['Date_and_Time_When_Order_Was_Placed'].notnull()].copy()  # Use .copy() to avoid the warning

# Extract the year from the date column using .loc
df_filtered.loc[:, 'year'] = df_filtered['Date_and_Time_When_Order_Was_Placed'].dt.year

# Group by 'Customer_Country', 'year', and 'Product_Category' and calculate total price
result = df_filtered.groupby(['Customer_Country', 'year', 'Product_Category']).agg(
    total_price=('Price', 'sum')  # Sum of 'Price' column
).reset_index()

# Sort the result
result = result.sort_values(by=['Customer_Country', 'year', 'total_price'], ascending=[True, True, False])

# Display the result
print(result)
```
